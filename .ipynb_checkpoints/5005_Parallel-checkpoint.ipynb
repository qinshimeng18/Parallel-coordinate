```python
%matplotlib inline
# %pylab inline
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_flow_aggregated_file.csv")
```

```python
g = df.groupby(['T1_Level2']).sum()
del g['transfer']
g['name'] = g.index
g.columns.values[0] = 'value'
g.to_json('T1_Level2.json',orient='records')
```
